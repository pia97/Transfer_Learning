```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("results.csv", sep=';')
```

```python
cnn = df[ df['model'] == 'CNN' ]
res = df[ df['model'] == 'RESNET' ]
inc = df[ df['model'] == 'INCEPTION' ]
enc = df[ df['model'] == 'ENCODER' ]
```

```python
transfer_sets = [ 'ArrowHead', 'Beef', 'BeetleFly', 'BirdChicken' ]
res_layers = [ 0, 13, 25, 37 ]
cnn_layers = [ 0, 4, 7, 11 ]
inc_layers = [ 0, 5, 11, 19, 22 ]
enc_layers = [ 0, 10, 19, 32, 41, 50, 63 ]
epochs = [ 20, 100, 500, 1000 ]
```

```python
def show_cnn_figure( set ):
    
    cnn_set = cnn[ cnn['pre_trained_set'] == set ]
    
    p = sns.barplot( x = 'epochs', y = 'accuracy', hue = 'num_frozen_layers', data = cnn_set, ci=0.95 ) 
    leg = p.get_legend()
    leg.set_title( "num_frozen_layers" )
    labs = leg.texts
    p.set_title( "CNN: " + set )

    p.axes.xaxis.label.set_text( "epochs" )
    p.axes.yaxis.label.set_text( "accuracy" )
```

```python
def show_architectures( set, epochs ):
    
    data = df[ df[ 'pre_trained_set' ] == set ]
    eps = data[ data[ 'epochs' ] == epochs ]
    p = sns.barplot( x = 'model', y = 'accuracy', data = eps, capsize = 0.05, errcolor = 'gray', errwidth = 2 ) 
    p.set_title( "architectures: " + set + " " + str( epochs ) )

    p.axes.yaxis.label.set_text( "accuracy" )
```

```python
for set in transfer_sets:
    
    show_architectures( set, 1000 )
    plt.show()
```

```python
for eps in epochs:
    show_architectures( 'ArrowHead', eps )
    plt.show()
```
